```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from pygal_maps_world.maps import World
```

## READING DATA FROM CSV FILE

```python
df = pd.read_csv('dataset.csv', usecols=list(range(1,23)), na_values="ND", index_col = 0, parse_dates = True)
# print('Dta')
df.head()
```

```python
print(df.columns)
# for col in df.columns:
#     x = col.split("-"[0].strip())[0]
#     df.col = x
#     print(col)
df.columns = [col.split("-")[0].strip() for col in df.columns]
print(df.columns)
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
df.iloc[0]
```

```python
scaled_df = df/df.iloc[0]
scaled_df.head()
```

```python
fig, ax = plt.subplots(figsize=(10,10))
plt.plot(scaled_df)
plt.xlabel('Year')
plt.title('Foreign Exchange Rate of Various Countries in the last 20 years')
_ = plt.legend(df.columns.values, bbox_to_anchor = (1,1))
```

## Part-1

```python
df[:"2010"]
```

```python
first_decade_df = df[:"2010"]
#inreasing here means exchange rate increasing, i.e, depreciation
increase_cols = [col for col in first_decade_df.columns if first_decade_df[col][-1]>first_decade_df[col][0]]
decrease_cols = [col for col in first_decade_df.columns if first_decade_df[col][-1]<first_decade_df[col][0]]

print("2000 to 2009")

first_decade_deps = []
print(str(len(increase_cols))+" countries' currencies have shown depreciation. They are:")
for i in increase_cols:
    print("\t" + i)
    first_decade_deps.append(i)

first_decade_apps = []
print(str(len(decrease_cols))+" countries' currencies have shown appreciation. They are:")
for i in decrease_cols:
    print("\t" + i)
    first_decade_apps.append(i)
```

```python
second_decade_df = df["2010":]
increase_cols = [col for col in second_decade_df.columns if second_decade_df[col][-1]>second_decade_df[col][0]]
decrease_cols = [col for col in second_decade_df.columns if second_decade_df[col][-1]<second_decade_df[col][0]]

print("2010 to 2019")

second_decade_deps = []
print(str(len(increase_cols))+" countries' currencies have shown depreciation. They are:")
for i in increase_cols:
    print("\t" + i)
    second_decade_deps.append(i)

second_decade_apps = []
print(str(len(decrease_cols))+" countries' currencies have shown appreciation. They are:")
for i in decrease_cols:
    print("\t" + i)
    second_decade_apps.append(i)
```

```python
increase_cols = [col for col in df.columns if df[col][-1]>df[col][0]]
decrease_cols = [col for col in df.columns if df[col][-1]<df[col][0]]

print("Overall")

overall_deps = []
print(str(len(increase_cols))+" countries' currencies have shown depreciation. They are:")
for i in increase_cols:
    print("\t" + i)
    overall_deps.append(i)

overall_apps = []
print(str(len(decrease_cols))+" countries' currencies have shown appreciation. They are:")
for i in decrease_cols:
    print("\t" + i)
    overall_apps.append(i)
```

```python
app_then_dep = []
dep_then_app = []
full_app = []
full_dep = []
for i in df.columns:
    if i in first_decade_apps and i in second_decade_apps:
        full_app.append(i)
    elif i in first_decade_apps and i in second_decade_deps:
        app_then_dep.append(i)
    elif i in first_decade_deps and i in second_decade_apps:
        dep_then_app.append(i)
    else:
        full_dep.append(i)

print(str(len(full_app)) + " regions appreciated continously:")
for i in full_app:
    print("\t" + i)
print(str(len(full_dep)) + " regions depreciated continously:")
for i in full_dep:
    print("\t" + i)
print(str(len(app_then_dep)) + " regions which appreciated then depriciated:")
for i in app_then_dep:
    print("\t" + i)
print(str(len(dep_then_app)) + " regions depriciated then appreciated:")
for i in dep_then_app:
    print("\t" + i)
```

## world map

```python
worldmap_chart = World()
country_codes = {
    'AUSTRALIA': 'au',
    'EURO AREA':['at', 'be', 'cy', 'ee', 'fi', 'fr', 'de', 'gr', 'ie', 'it', 'lv', 'lt', \
                'lu', 'mt', 'nl', 'pt', 'sk', 'si', 'es'],
    'NEW ZEALAND':'nz',
    'UNITED KINGDOM':'uk',
    'BRAZIL':'br',
    'CANADA':'ca',
    'CHINA':'ch',
    'HONG KONG':'hk',
    'INDIA':'in',
    'KOREA':['kp', 'kr'],
    'MEXICO':'mx',
    'SOUTH AFRICA':'za',
    'SINGAPORE':'sg',
    'DENMARK':'dk',
    'JAPAN':'jp',
    'MALAYSIA':'my',
    'NORWAY':'no',
    'SWEDEN': 'se',
    'SRI LANKA': 'lk',
    'SWITZERLAND': 'ch',
    'TAIWAN': 'tw'
}
worldmap_chart.title = 'Percentage Appreciation and Depreciation of currencies over the last 20 years'
dic = {}
for col in scaled_df.columns:
    code = country_codes[col]
    change = scaled_df[col][-1]
    if type(code)!=list:
        dic[code] = change
    else:
        for i in code:
            dic[i] = change
neg_dic = {}
pos_dic = {}
for code, change in dic.items():
    if change < 1.0:
        neg_dic[code] = round((1.0-change)*100, 2)
    else:
        pos_dic[code] = round((change-1.0)*100, 2)
worldmap_chart.add("Depreciation", pos_dic)
worldmap_chart.add("Appreciation", neg_dic)
worldmap_chart.render_to_file("WorldMap.svg")
```

## Part 4

```python
#week volatality
vols = {}
for col in df.columns:
    vols[col] = scaled_df.groupby([scaled_df.index.strftime('%Y'), scaled_df.index.strftime('%W')])[col].max() - \
        scaled_df.groupby([scaled_df.index.strftime('%Y'), scaled_df.index.strftime('%W')])[col].min()
vols = pd.DataFrame(vols)
volmeans = vols.mean()
plt.bar(df.columns, volmeans)
plt.title("Weekly Volatility oF Foreign Exchange Rates")
a = plt.xticks(rotation = 90)
```
